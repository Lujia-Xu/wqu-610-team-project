### 0.Data read

```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
from sklearn.linear_model import LassoCV, RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

path = 'data/FE-GWP1_model_selection_2.csv'
df = pd.read_csv(path)
df.columns = df.columns.str.strip()
df.head()
```

### 1. Multiple Linear Regression Model

**Model Fit Metrics:**
- **Adjusted R²:** 0.9936  
- **AIC:** -165.90  
- **BIC:** -150.27

```python
X = df[['Z1','Z2','Z3','Z4','Z5']]
y = df['Y']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

print(model.summary())
print("\nAdj_R2:", model.rsquared_adj)
print("AIC:", model.aic)
print("BIC:", model.bic)
```

### 2. LASSO Regression Model

We fit a LASSO regression with cross-validation to select the optimal regularization parameter.

**Model Fit Metrics:**
- **Adjusted R²:** 0.99357  
- **AIC:** -165.81  
- **BIC:** -150.18

```python
X = df[['Z1','Z2','Z3','Z4','Z5']].values
y = df['Y'].values

model = make_pipeline(StandardScaler(), LassoCV(cv=5, random_state=0))
model.fit(X, y)

lasso = model.named_steps['lassocv']
print("alpha:", lasso.alpha_)

coef_df = pd.DataFrame({'var':['Z1','Z2','Z3','Z4','Z5'],'coef':lasso.coef_})
print(coef_df)

y_hat = model.predict(X)
n = len(y)
rss = ((y - y_hat)**2).sum()
tss = ((y - y.mean())**2).sum()
r2 = 1 - rss / tss
k = (np.abs(lasso.coef_) > 1e-8).sum() + 1
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k)
sigma2 = rss / n
loglik = -0.5 * n * (np.log(2 * np.pi * sigma2) + 1)
aic = 2 * k - 2 * loglik
bic = np.log(n) * k - 2 * loglik

print("R2:", r2)
print("Adj_R2:", adj_r2)
print("AIC:", aic)
print("BIC:", bic)
```

### 3. Ridge Regression Model

Ridge regression was fitted with cross-validation over log-spaced alpha values.  
The model applies L2 regularization, shrinking coefficients smoothly without performing variable elimination.

**Model Fit Metrics:**
- **Adjusted R²:** 0.99356  
- **AIC:** -165.66  
- **BIC:** -150.03

```python
X = df[['Z1','Z2','Z3','Z4','Z5']].values
y = df['Y'].values

alphas = np.logspace(-4, 4, 50)
ridge_model = make_pipeline(StandardScaler(), RidgeCV(alphas=alphas, cv=5))
ridge_model.fit(X, y)

ridge = ridge_model.named_steps['ridgecv']
print("alpha:", ridge.alpha_)

coef_df_ridge = pd.DataFrame({'var':['Z1','Z2','Z3','Z4','Z5'],'coef':ridge.coef_})
print(coef_df_ridge)

y_hat = ridge_model.predict(X)
n = len(y)
rss = ((y - y_hat)**2).sum()
tss = ((y - y.mean())**2).sum()
r2 = 1 - rss / tss
k = 5 + 1
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - k)
sigma2 = rss / n
loglik = -0.5 * n * (np.log(2 * np.pi * sigma2) + 1)
aic = 2 * k - 2 * loglik
bic = np.log(n) * k - 2 * loglik

print("R2:", r2)
print("Adj_R2:", adj_r2)
print("AIC:", aic)
print("BIC:", bic)
```
